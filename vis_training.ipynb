```python
import json
import matplotlib.pyplot as plt
```

```python
def get_values_excluding_push(stats):
    return [s['value'] for s in stats if s['epoch_type'] not in ['PUSH']]
```

```python
ds_name = 'ArticularyWordRecognition'
```

```python
exp = f'{ds_name}/TestExperimentScalingNonReg'
for i in range(1, 6):
    with open(f'./experiments/{exp}{i}/stats.json') as f:
        stats1 = json.load(f)
    plt.plot(get_values_excluding_push(stats1['cross_ent_test']), label=exp)
plt.legend()
```
